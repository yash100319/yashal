17. What are point estimates and interval estimates in statistics ?
-  1. Point Estimate
A point estimate is a single value (a specific number) that serves as the best guess or approximation for an unknown population parameter. It is based on sample data and does not provide any information about the possible variability or uncertainty of that estimate.

Example: If you take a sample of 100 people and calculate the sample mean age to be 35 years, this 35 years is a point estimate of the population mean age.

Characteristics:

Provides a single value.

No measure of uncertainty or variability is given with a point estimate.

2. Interval Estimate
An interval estimate, on the other hand, provides a range of values within which the population parameter is likely to lie, along with a measure of uncertainty. It gives not just an estimate but also an associated confidence level, usually expressed as a confidence interval.

Example: Instead of just saying the sample mean is 35 years, you might say, "We are 95% confident that the true population mean age lies between 33 and 37 years." This range, from 33 to 37 years, is an interval estimate.

Characteristics:

Provides a range of values.

Includes a level of confidence (like 95% or 99%), indicating how confident we are that the true parameter lies within the interval.
18. What is the significance of confidence intervals in statistical analysis ?
- 1. Estimation of Population Parameters
Instead of giving a single point estimate (like the sample mean), a confidence interval provides a range that is likely to contain the true population parameter. This helps in understanding the uncertainty around the estimate.

2. Quantifying Uncertainty
In real-world data, there is always some degree of uncertainty due to sampling variability. A confidence interval expresses this uncertainty by showing the range within which the true parameter is expected to lie, with a certain level of confidence (e.g., 95%, 99%).

3. Interpretation of Confidence Level
The confidence level (e.g., 95%) indicates how confident we are that the interval contains the true parameter. If you were to take 100 different samples and calculate 100 different confidence intervals, 95 of those intervals would contain the true population parameter, assuming a 95% confidence level.

21. What are the assumptions for applying the Central Limit Theorem ?
- The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes the distribution of the sample mean (or sum) for a large enough sample size. For the CLT to apply, several key assumptions are typically made:

Independence: The observations in the sample must be independent of each other. This means that the selection of one observation should not affect the selection of another. In other words, the data points should come from different individuals or units that are unrelated.

Random Sampling: The sample should be drawn randomly from the population. This ensures that the sample is representative of the population, minimizing bias.

Sample Size: The sample size should be sufficiently large. While there‚Äôs no strict rule, a common guideline is that the sample size should be at least 30. However, if the population is highly skewed or has extreme outliers, a larger sample size may be required to approximate a normal distribution.

Finite Variance: The population from which the sample is drawn should have a finite variance. If the variance is infinite, the CLT does not hold.

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution


```python
def bernoulli_pdf(x, p):
    """
    Calculate the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The value of the random variable (0 or 1).
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability P(X = x) based on the Bernoulli distribution.
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("x must be either 0 or 1.")

# Example usage:
p = 0.7  # Probability of success
print(bernoulli_pdf(1, p))  # Probability of success (1)
print(bernoulli_pdf(0, p))  # Probability of failure (0)
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution?


```python
import numpy as np
import scipy.stats as stats

def calculate_z_score(data, value):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-score
    z_score = (value - mean) / std_dev
    return z_score

def compare_to_normal_distribution(z_score):
    # Compare Z-score to standard normal distribution
    p_value = stats.norm.cdf(z_score)  # CDF gives probability less than the Z-score
    return p_value

# Example usage
data = [23, 25, 27, 30, 31, 35, 40, 42, 44, 45]  # Example data set
value = 30  # Data point we are interested in

z_score = calculate_z_score(data, value)
print(f"Z-score for value {value}: {z_score}")

# Compare the Z-score to the standard normal distribution
probability = compare_to_normal_distribution(z_score)
print(f"Probability (CDF) for Z-score {z_score}: {probability:.4f}")
```

15. Implement hypothesis testing using Z-statistics for a sample dataset ?


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = np.array([120, 122, 119, 124, 126, 121, 123, 118, 125, 127])

# Known population parameters
population_mean = 120  # ¬µ0
population_std = 5     # œÉ (standard deviation)
alpha = 0.05          # Significance level

# Step 1: Compute sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Step 2: Calculate the Z-statistic
Z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 3: Calculate the p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(Z)))

# Step 4: Decision rule
if p_value < alpha:
    result = "Reject the null hypothesis"
else:
    result = "Fail to reject the null hypothesis"

# Output results
print(f"Sample Mean: {sample_mean}")
print(f"Z-statistic: {Z}")
print(f"P-value: {p_value}")
print(result)
```

18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution ?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the mean and standard deviation
mean = 0
std_dev = 1

# Create an array of values from -5 to 5
x = np.linspace(-5, 5, 1000)

# Calculate the PDF of the normal distribution
pdf = norm.pdf(x, mean, std_dev)

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Normal Distribution\nMean={mean}, Std Dev={std_dev}', color='blue')

# Add labels and title
plt.title('Probability Density Function (PDF) of Normal Distribution', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)

# Add grid
```
